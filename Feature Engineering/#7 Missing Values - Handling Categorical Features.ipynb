#### One Hot Encoding with many categories in a feature

```python
df = pd.read_csv('mercedesbenz.csv')
```

```python
df.head()
```

```python
df = pd.read_csv('mercedesbenz.csv', usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

#### Use only the top 10 most occuring feature for encoding and remove the rest of the features

```python
lst_10 = df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10 = list(lst_10)
```

```python
lst_10
```

```python
import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
lst_10.append('X1')
```

```python
df[lst_10]
```
